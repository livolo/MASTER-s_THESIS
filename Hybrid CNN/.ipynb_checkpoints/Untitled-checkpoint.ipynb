```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
#from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications import EfficientNetB2
from tensorflow.keras.applications.efficientnet import preprocess_input
from sklearn.model_selection import train_test_split
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
import tensorflow as tf
print("Devices:", tf.config.list_physical_devices())
```

```python
IMG_SIZE = 260

def load_image(path):
    img = load_img(path, target_size=(IMG_SIZE, IMG_SIZE))
    img = img_to_array(img)
    return preprocess_input(img)

# Load metadata
df = pd.read_csv("Breastcancer.csv")  # filename, age, birads, density, label

# Convert folder + filename → full path
df["path"] = "/home/rgitbt/sagar/codebook/dataset/" +  df["File_Name"]

# Load all images into numpy array
X_images = np.array([load_image(p) for p in df["path"]])

# Clinical variables
X_clinical = df[["Age", "BIRADS category", "Breast_density_category"]].values.astype("float32")

# Labels
y = df["Label"].values
```

```python
#Initial Train–Test Split (80–20)
Ximg_trainval, Ximg_test, Xclin_trainval, Xclin_test, y_trainval, y_test = train_test_split(
    X_images, X_clinical, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

#rain–Validation Split (80–20 of remaining)
Ximg_train, Ximg_val, Xclin_train, Xclin_val, y_train, y_val = train_test_split(
    Ximg_trainval, Xclin_trainval, y_trainval,
    test_size=0.20,
    random_state=42,
    stratify=y_trainval
)
```

```python
def build_hybrid_model():
    # Image branch
    img_input = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    base = EfficientNetB2(include_top=False, weights="imagenet", pooling="avg")
    base.trainable = False   # freeze CNN

    x = layers.GlobalAveragePooling2D()(base.output)
    x = layers.Dense(256, activation="relu")(x)

    # Clinical branch
    clin_input = layers.Input(shape=(3,))
    c = layers.Dense(32, activation="relu")(clin_input)

    # Fusion
    combined = layers.concatenate([x, c])
    combined = layers.Dense(64, activation="relu")(combined)
    output = layers.Dense(1, activation="sigmoid")(combined)

    model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        metrics.AUC(name='auc'),
        metrics.Precision(name='precision'),
        metrics.Recall(name='recall'),
        metrics.TruePositives(name='tp'),
        metrics.FalsePositives(name='fp'),
        metrics.TrueNegatives(name='tn'),
        metrics.FalseNegatives(name='fn')
    ]
    )
    model.summary()
    return model
```

```python
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.callbacks import ModelCheckpoint
import numpy as np

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold = 1
auc_scores = []

for train_idx, val_idx in kfold.split(X_images, y):
    print(f"\n🔁 Fold {fold}")

    Ximg_train, Ximg_val = X_images[train_idx], X_images[val_idx]
    Xclin_train, Xclin_val = X_clinical[train_idx], X_clinical[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    model = build_hybrid_model(
        img_shape=X_images.shape[1:],
        clin_shape=X_clinical.shape[1]
    )

    checkpoint = ModelCheckpoint(
        f"best_model_fold{fold}.h5",
        monitor="val_auc",
        mode="max",
        save_best_only=True,
        verbose=1
    )

    history = model.fit(
        [Ximg_train, Xclin_train],
        y_train,
        validation_data=([Ximg_val, Xclin_val], y_val),
        epochs=20,
        batch_size=16,
        callbacks=[checkpoint],
        verbose=1
    )

    val_auc = max(history.history["val_auc"])
    auc_scores.append(val_auc)

    print(f"Fold {fold} best AUC: {val_auc:.4f}")
    fold += 1


```

```python
import numpy as np
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    f1_score,
    matthews_corrcoef
)

# Predict probabilities
y_prob = model.predict([Ximg_test, Xclin_test])
y_pred = (y_prob > 0.5).astype(int)

# Confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

# Metrics
specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

print("\n--- Test Metrics ---")
print(f"Accuracy     : {(tp + tn) / (tp + tn + fp + fn):.4f}")
print(f"AUC          : {auc:.4f}")
print(f"Sensitivity  : {sensitivity:.4f}")
print(f"Specificity  : {specificity:.4f}")
print(f"Precision    : {tp / (tp + fp):.4f}")
print(f"F1-score     : {f1:.4f}")
print(f"MCC          : {mcc:.4f}")
```

```python
#BOOTSTRAP CI FOR AUC
def bootstrap_auc(y_true, y_prob, n_bootstraps=2000):
    rng = np.random.RandomState(42)
    scores = []

    for _ in range(n_bootstraps):
        idx = rng.randint(0, len(y_true), len(y_true))
        if len(np.unique(y_true[idx])) < 2:
            continue
        scores.append(roc_auc_score(y_true[idx], y_prob[idx]))

    return np.percentile(scores, [2.5, 97.5])

ci_low, ci_high = bootstrap_auc(y[test_idx], y_prob)
print(f"AUC 95% CI: [{ci_low:.3f}, {ci_high:.3f}]")
```
